# Gr√°fico interativo no Python WEGE3 e ABEV3

```python
import yfinance as yf
import plotly.graph_objs as go
```

```python
tickers = ['ABEV3.SA', 'WEGE3.SA']
data = yf.download(tickers, '2010-12-31')['Adj Close']

traces = []
for ticker in tickers:
    traces.append(go.Scatter(x = data.index, y = data[ticker], mode = 'lines', name = ticker))
    
layout = go.Layout(xaxis_title = 'Data', yaxis_title = 'Pre√ßo')
fig = go.Figure(data = traces, layout = layout)
fig.show()
```
