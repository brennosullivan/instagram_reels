# Modelo de média móvel Python + Pandas

```python
import yfinance as yf
import pandas_ta as ta
import matplotlib.pyplot as plt
```

```python
data = yf.download("WEGE3.SA",start="2015-01-01")
data['SMA 30'] = ta.sma(data['Close'],30)
data['SMA 100'] = ta.sma(data['Close'],100)

data
```

```python
fig, ax = plt.subplots(figsize=(14,8))
ax.plot(data['Adj Close'] , label = "WEGE3.SA" ,linewidth=0.8, color='blue', alpha = 0.9)
ax.plot(data['SMA 30'], label = 'SMA30', alpha = 0.85,linewidth=0.9)
ax.plot(data['SMA 100'], label = 'SMA100' , alpha = 0.85,linewidth=0.9)

plt.show()
```
