# Como pegar taxas do Tesouro Direto com Python

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
driver = webdriver.Chrome(service = Service(ChromeDriverManager().install()))

url = '''https://www.tesourodireto.com.br/titulos/precos-e-taxas.htm'''

driver.get(url)
```

```python
#trocar pra visualização de tabela

trocar_pra_tabela = driver.find_element("xpath",  '//*[@id="switchVerTabelas"]')

#clicar no botao

driver.execute_script("arguments[0].click();", trocar_pra_tabela)

#achando a tabela de taxas
tabela_taxas = driver.find_element("xpath", '//*[@id="td-precos_taxas-tab_1"]/div/div[10]/table')

#pegando HTML

html_tabela = tabela_taxas.get_attribute("outerHTML")

#lendo html

tabela_final = pd.read_html(html_tabela)[0][['Título', 'Rentabilidade anual:',
                                            'Investimento mínimo:', 'Preço Unitário:',
                                            'Vencimento:']]

#fechando navegador e mostrando a tabela

driver.quit()

tabela_final
```
