```python
dados_receitas = pd.read_csv("receitas_candidatos_2022_BRASIL.csv", 
                             sep = ";", 
                             encoding = "latin-1",
                            usecols = ['NM_CANDIDATO', 'SG_PARTIDO', 
                             'NM_DOADOR', 'VR_RECEITA', 'NR_CPF_CNPJ_DOADOR'],
                            decimal = ",")
dados_receitas
```

```python
lista_dfs = []

doadores = ["SALIM MATTAR", "MARCOS ERMINIO DE MORAES", 'HEITOR VANDERLEI LINDEN',
            'ALEXANDRE GRENDENE BARTELLE', 'PEDRO GREDENE BARTELLE', 'RUBENS OMETTO',
            'CANDIDO BOTELHO BRACHER', 'ARMINIO FRAGA NETO', 'ODILIO BALBINOTTI FILHO',
            'NELSON PIQUET', 'WALTER MOREIRA SALLES', 
            'JOÃO MOREIRA SALLES', 'OSCAR LUIZ CERVI', 'ROBERT CARLOS LYRA']
              

cpf = [7182376687, 77232682620, 23613068087, 9867597087, 9864784072, 41232178853, 3969018838,
      46906525700, 59677015915, 76168328149, 40693546700, 66719739700, 21062803000,
      13381334468]


df_doador = dados_receitas[dados_receitas['NR_CPF_CNPJ_DOADOR'].isin(cpf)]

df_doador

for i, doador in enumerate(doadores):
    
    df_doador.loc[df_doador['NR_CPF_CNPJ_DOADOR'] == cpf[i], 'NM_DOADOR'] = doador
```

```python
df_por_partido = df_doador.groupby(['SG_PARTIDO', 'NM_DOADOR'])['VR_RECEITA'].agg('sum')

df_por_partido.to_excel('doacoes_por_pessoa_para_partidos.xlsx')
```
