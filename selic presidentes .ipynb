```python
import matplotlib.pyplot as plt
from bcb import sgs
from datetime import datetime
import mplcyberpunk


plt.style.use("cyberpunk")

# taxa selic definida pelo Copom

selic = sgs.get({'SELIC':432}, start='2003-01-01').resample('M').last()


data_final_mandatos =[(datetime(2010,12,31),' Fim\n Lula 2'),
                      (datetime(2015,12,31),' Fim\n Dilma 2'),
                      (datetime(2018,12,31),' Fim\n Temer'),
                      (datetime(2022,12,31),' Fim\n Bolsonaro')]

fig, ax = plt.subplots(figsize = (10,7.5))
ax.plot(selic)

for data, evento in data_final_mandatos:

    ax.annotate(evento, xy= (data,selic['SELIC'].asof(data) + 1),
                        xytext= (data, selic['SELIC'].asof(data) + 4),
                        arrowprops = dict(facecolor = 'white', 
                                          edgecolor = "white", headwidth = 5, width = 2, headlength = 5),
                        horizontalalignment = 'center', verticalalignment = 'top')

plt.title("Taxa Selic (%) a cada mandato", color = "w")

plt.savefig('selicpresidentes.png', dpi = 300)
    
plt.show()
```
