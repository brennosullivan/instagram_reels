```python
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
cotacoes = yf.download(["^BVSP", "USDBRL=X"], "2015-12-31")['Close']

cotacoes = cotacoes.resample("Y").last().pct_change().dropna()

cotacoes['ano'] = cotacoes.index.year
cotacoes['ano'] = cotacoes['ano'].astype("category")

cotacoes.columns = ['Ibovespa', 'Dólar', 'Ano']

scatter_plot = cotacoes.plot.scatter(x = "Dólar", y = "Ibovespa", c = "Ano", cmap = 'viridis')

scatter_plot.axhline(y = 0, color = "black", linewidth = 2)
scatter_plot.axvline(x = 0, color = "black", linewidth = 2)

plt.show()
```
