```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.catho.com.br/vagas/cientista-de-dados/?q=Cientista%20de%20Dados"
response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')
data = {'job_title': [], 'company': [], 'location': [], 'salary': [], 'description': []}

job_titles = soup.find_all("h2",{"class":'Title__Heading-sc-14fvmc0-0 fGTSAd sc-hwwEjo kPUpWf'})
companies = soup.find_all("p",{"class":'sc-kPVwWT gyAXcO'})
locations = soup.find_all("button",{"class":'sc-hXRMBi ewWuRq'})
salaries = soup.find_all("div",{"class":'sc-hZSUBg bcovgZ'})
descriptions = soup.find_all("span",{"class":'job-description'})

for title, company, location, salary, description in zip(job_titles, companies, locations, salaries, descriptions):
    data['job_title'].append(title.text)
    data['company'].append(company.text)
    data['location'].append(location.text)
    data['salary'].append(salary.text.replace("De R$ ", ""))
    data['description'].append(description.text)

df = pd.DataFrame(data)
df
```
