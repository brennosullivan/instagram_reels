# Performance acumulada WEGE3, PETR4, VALE3 e LREN3

```python
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
df = yf.download(['WEGE3.SA', "PETR4.SA", "VALE3.SA", "LREN3.SA"], "2010-12-31")['Adj Close']

df = ((df.pct_change().dropna()) + 1).cumprod() - 1

df_anual = df.resample("Y").last()

df_anual.index = df_anual.index.strftime("%Y")

df_anual.plot.bar(stacked = True)

plt.show()
```
