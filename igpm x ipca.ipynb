```python
import matplotlib.pyplot as plt
from bcb import sgs
import matplotlib.ticker as ticker
import mplcyberpunk
from matplotlib.dates import date2num

plt.style.use("cyberpunk")

dados = sgs.get({'IPCA':433, 'IGPM':189}, start='2022-06-01').resample('M').last()

datas_numericas = date2num(dados.index)

fig, ax = plt.subplots(figsize = (10, 7.5))

ax.bar(datas_numericas-7, dados['IPCA'], label = "IPCA", width=7)
ax.bar(datas_numericas, dados['IGPM'], label = "IGP-M", width=7)

ax.yaxis.set_major_formatter(ticker.PercentFormatter())
ax.xaxis_date()
ax.grid(False)
plt.axhline(y=0, color = 'w')
plt.legend()
```
